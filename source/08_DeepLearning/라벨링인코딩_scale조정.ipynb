# 전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv (지역명, 평당분양가격, 연도, 월)
- 지역명2 : 지역명필드를 라벨인코딩하여 추가
- 평당분양가격n : 평당분양가격의 정규화(normalization) 스케일 조정하여 추가
- 평당분양가격s : 평당분양가격의 표준화(standardization) 스케일 조정하여 추가
- 연도n : 연도의 정규화(normalization) 스케일 조정하여 추가
- 연도s : 연도의 표준화(standardization) 스케일 조정하여 추가
- 월n : 월의 정규화(normalization) 스케일 조정하여 추가
- 월s : 월의 표준화(standardization) 스케일 조정하여 추가

```python
import pandas as pd
df = pd.read_csv('D:\imsehyeon\Downloads\shareData\전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv', encoding='cp949')
df.head()
```

```python
df = df.melt(id_vars='지역', var_name='연도월', value_name='평당분양가격')
df['연도'] = df['연도월'].apply(lambda x : int(x.split('년')[0]))
df['월'] = df['연도월'].apply(lambda x : int(x.split('년')[-1].replace('월', '')))
df.drop('연도월', axis=1, inplace=True)
df = df.rename(columns={'지역' : '지역명'})
df.head()
```

```python
# 열별 스케일 조정
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
import numpy as np
le = LabelEncoder()
```

```python
# 지역명2 : 지역명필드를 라벨인코딩하여 추가
df['지역명2'] = le.fit_transform(df['지역명'])
df.head()
```

# 정규화 스케일

```python
# # 평당분양가격n : 평당분양가격의 정규화 스케일 조정하여 추가 (라이브러리 미사용)
# df['평당분양가격n'] = (df['평당분양가격'] - df['평당분양가격'].min()) / (df['평당분양가격'].max() - df['평당분양가격'].min())

# # 연도n : 연도의 정규화 스케일 조정하여 추가 (라이브러리 미사용)
# df['연도n'] = (df['연도'] - df['연도'].min()) / (df['연도'].max() - df['연도'].min())

# # 월n : 월의 정규화 스케일 조정하여 추가 (라이브러리 미사용)
# df['월n'] = (df['월'] - df['월'].min()) / (df['월'].max() - df['월'].min())
# df.head()
```

```python
# 평당분양가격n : 평당분양가격의 정규화 스케일 조정하여 추가 (라이브러리 사용)
price_data = np.array(df['평당분양가격']).reshape(-1, 1)
price_scaler = MinMaxScaler()
price_scaler.fit(price_data)
df['평당분양가격n'] = price_scaler.transform(price_data)

# 연도n : 연도의 정규화 스케일 조정하여 추가 (라이브러리 사용)
year_data = np.array(df['연도']).reshape(-1, 1)
#         df['연도'].values.reshape(-1, 1)
year_scaler = MinMaxScaler()
year_scaler.fit(year_data)
df['연도n'] = year_scaler.transform(year_data)  # transform 하기 위해서는 2차원 데이터가 들어가야 하므로 n2_data에 reshape(-1, 1) 해줌
# df['연도n1'] = scaler_n2.fit_transform(np.array(df['연도']).reshape(-1, 1))
# df['연도n1'] = scaler_n2.fit_transform(df['연도'].values.reshape(-1, 1))
# df['연도n1'] = scaler_n2.fit_transform(df[['연도']].values)  # 한 줄로 : 위의 세 줄의 결과는 동일함

# 월n : 월의 정규화 스케일 조정하여 추가 (라이브러리 사용)
month_data = np.array(df['월']).reshape(-1, 1)
month_scaler = MinMaxScaler()
month_scaler.fit(month_data)
df['월n'] = month_scaler.transform(month_data)
df.head()
```

# 표준화 스케일

```python
# # 평당분양가격s : 평당분양가격의 표준화 스케일 조정하여 추가 (라이브러리 미사용)
# df['평당분양가격s'] = (df['평당분양가격'] - df['평당분양가격'].mean()) / df['평당분양가격'].std()
# # 연도s : 연도의 표준화 스케일 조정하여 추가 (라이브러리 미사용)
# df['연도s'] = (df['연도'] - df['연도'].mean()) / df['연도'].std()
# # 월s : 월의 표준화 스케일 조정하여 추가 (라이브러리 미사용)
# df['월s'] = (df['월'] - df['월'].mean()) / df['월'].std()
# df.head()
```

```python
# 평당분양가격s : 평당분양가격의 표준화 스케일 조정하여 추가 (라이브러리 사용)
price_data = np.array(df['평당분양가격']).reshape(-1, 1)
price_scaler = StandardScaler()
price_scaler.fit(price_data)
df['평당분양가격s'] = price_scaler.transform(price_data)

# 연도s : 연도의 표준화 스케일 조정하여 추가 (라이브러리 사용)
year_data = np.array(df['연도']).reshape(-1, 1)
year_scaler = StandardScaler()
year_scaler.fit(year_data)
df['연도s'] = year_scaler.transform(year_data)

# 월s : 월의 표준화 스케일 조정하여 추가 (라이브러리 사용)
s3_data = np.array(df['월']).reshape(-1, 1)
scaler_s3 = StandardScaler()
scaler_s3.fit(s3_data)
df['월s'] = scaler_s3.transform(s3_data)
df.head()
```
