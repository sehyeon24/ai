```python
nanex = ex.loc[368]   # 결측치인 행
spaceex = ex.loc[28]  # 공백인 행
nanex, spaceex, ex.loc[0]
```

```python
# 분양가격이 nan이거나 space면 0, 그 외 분양가격의 문자자리수를 return하는 함수
def get_pricelen(row):  # row : 한 행  : 시리즈
    price = row['분양가격(제곱미터)']
    # print(row.notnull()['분양가격(제곱미터)'])  # nownull() : 결측치가 아니면 True, 결측치이면 False
    if row.notnull()['분양가격(제곱미터)'] and price.strip() != '':
        return len(price)
    else:
        return 0
get_pricelen(ex.loc[0]), get_pricelen(nanex), get_pricelen(spaceex)
```

```python
# get_pricelen 함수를 이용해 분양가격(제곱미터)의 자릿수를 나타내는 가격자리수 컬럼 생성
ex['가격자리수'] = ex.apply(get_pricelen, axis=1)
ex.head()
```

### apply

```python
# 콤마(,)가 있는 문자에서 콤마 없애기 방법1
temp = '12,523'
tempsplited = temp.split(',')
print(tempsplited)
print(''.join(tempsplited))
```

```python
# 콤마(,)가 있는 문자에서 콤마 없애기 방법2
temp = '12,523'
print(temp.replace(',', ''))
```

```python
# 분양가격 컬럼에 콤마(,)가 있는 경우 콤마(,)를 제거('5,123' -> '5123')
#              공백문자가 있는 경우 np.nan로 바꾸는 함수
def commadrop(row):
    # row = row.copy()  # 에러메세지 없애기 위함
    # pd.options.mode.copy_on_write = True  # 에러메세지 없애기 위함 # 데이터 복사를 최소화
    price = row['분양가격(제곱미터)']
    if row.notnull()['분양가격(제곱미터)']:  # 분양가격이 결측치가 아닌 경우
        if price.find(',') != -1:  # 콤마(,) 없애기
            # row['제곱미터당분양가격'] = price.replace(',', '')
            pricesplited = price.split(',')
            row['제곱미터당분양가격'] = ''.join(pricesplited)
        elif price.strip() == '':
            row['제곱미터당분양가격'] = np.nan
        else:
            row['제곱미터당분양가격'] = price
    else:  # 분양가격이 결측치인 경우
        row['제곱미터당분양가격'] = np.nan
    return row

commadrop(df_last.loc[0])  # 콤마 제외
commadrop(df_last.loc[1])  # 그대로
commadrop(df_last.loc[28])  # 공백문자의 경우 nan
commadrop(df_last.loc[368])  # nan의 경우 nan
```

```python
# 공백문자는 결측치로, 콤마는(,)는 제거하기를 모든 행에 적용
df_last = df_last.apply(commadrop, axis=1)
df_last.head(3)
```

```python
# df_last.drop(['평당분양가격', '제곱미터당분양가격'], axis=1, inplace=True)
df_last.sample(1)
```

```python
# 결측치 -> 결측치 return, 공백문자 -> 결측치 return, '3,198' -> 3198 return, '3198' -> 3198 return
def commadrop2(price):
    # np.isnan(값) : nan만 결측치로 인식하고 결측치인지 여부를 return
    # pd.isna(값), pd.isnull(값) : nan과 None을 모두 결측치로 인식하고 결측치인지 여부를 return
    # 값 is np.nan : 값이 nan인지 여부를 return
    if price is np.nan:  # 결측치인지
        return np.nan
    elif price.strip() == '':  # 공백문자인지
        return np.nan
    elif price.find(',') != -1:  # 콤마(,) 여부
        pricesplited = price.split(',')
        price = ''.join(pricesplited)
        return int(price)
    else:
        return int(price)
commadrop2(df_last.loc[0, '분양가격(제곱미터)'])  # 콤마를 제외한 문자 -> 숫자
commadrop2(df_last.loc[1, '분양가격(제곱미터)'])  # 콤마없는 문자 -> 숫자
commadrop2(df_last.loc[28, '분양가격(제곱미터)'])  # 공백문자 -> nan
commadrop2(df_last.loc[368, '분양가격(제곱미터)'])  # nan -> nan
```

- apply(함수, axis) : 1차원, 2차원(axis 지정. axis 기본값 0)
- map(함수) : 1차원에서만 사용 가능
- applymap(함수) : 2차원(모든 요소들에 함수 적용)

```python
# df_last['분양가격(제곱미터)'].apply(commadrop2)  # apply 앞이 2차원일 때 axis 필요 / apply 앞이 1차원일 때는 axis 필요 X
df_last['제곱미터당분양가격'] = df_last['분양가격(제곱미터)'].map(commadrop2)
df_last.tail(6)                    
```

## ① 지역별 데이터 수 (cf. 평당분양가격이 있는 데이터 수)

```python
# 지역명별 데이터 행의 갯수(결측치 포함)
df_last['지역명'].value_counts()  # 정렬
```

```python
# 지역명별 데이터 행의 갯수(결측치 제외) : 결과가 1차원 시리즈
df_last.groupby(['지역명'])['평당분양가격'].count().sort_values(ascending=False)  # 내림차순
```

```python
# 지역명별 데이터 행의 갯수(결측치 제외) : 결과가 2차원 데이터프레임
# 1.
# df_last.groupby(['지역명'])['평당분양가격'].count().to_frame().sort_values(by='평당분양가격', ascending=False)
# 2.
# df_last.groupby(['지역명'])['평당분양가격'].count().reset_index()
# 3.
# df_last.groupby(['지역명'], as_index=False)['평당분양가격'].count()
# 4.
df_last.groupby(['지역명'])[['평당분양가격']].count()
```

```python
# 지역명별 결측치를 제외한 데이터 수(pivot_table()이용 : 결과는 2차원 데이터프레임)
df_last.pivot_table(index=['지역명'],
                    values=['평당분양가격'],
                    aggfunc='count')
pd.pivot_table(df_last, index=['지역명'],
                    values=['평당분양가격'],
                    aggfunc='count').T
pd.pivot_table(df_last, index='지역명',
                    values='평당분양가격',
                    aggfunc='count').transpose()
```

## ④ 지역별, 전용면적별 평당분양가격 (평균)

```python
# 1차원 시리즈
r = df_last.groupby(['지역명', '전용면적'])['평당분양가격'].mean()
print(r)
print()
print('지역명 중 서울만 : \n', r['서울'])
print()
# 서울 중 '모든 면적'의 평균
print('서울의 모든 면적의 평균 : ', r['서울']['모든면적'].mean())
print('서울의 모든 면적의 평균 : ', r['서울', '모든면적'].mean())
```

```python
# 2차원 데이터프레임 : to_frame(), reset_index()
r = df_last.groupby(['지역명', '전용면적'])[['평당분양가격']].mean()
display(r)
# 서울만
display(r.loc['서울'])
# 서울 중 모든면적
display(r.loc['서울', '모든면적'])
display(r.loc['서울'].loc['모든면적'])
```

```python
# level=0인 index는 그대로 두고, level=1인 index를 컬럼으로 적용
# 지역명을 행이름으로, 전용면적(끝 index)을 열이름으로
df_last.groupby(['지역명', '전용면적'])['평당분양가격'].mean().unstack()
```

```python
df_last.pivot_table(index=['지역명', '전용면적'],
                    values='평당분양가격',
                    # aggfunc='mean'
                   ).unstack()
```

```python
r = df_last.pivot_table(index='지역명',
                    columns='전용면적',
                    values='평당분양가격',
                    aggfunc='mean')
r = r[['60㎡', '60㎡~85㎡', '85㎡~102㎡', '102㎡~', '모든면적']]
r
```
