## ③ 결측치를 확인하고 결측치가 없는 컬럼을 포함하여 시각화(정렬전 bar plot, barh plot, 정렬 후 bar plot, barh plot)

```python
# 열별 결측치 갯수 확인
missing_cnt = df.isna().sum()  # axis=0 생략으로 열별 합계
missing_cnt
```

```python
# 결측치를 세로 막대그래프(bar)로 시각화
# missing_cnt.plot(kind='bar')
missing_cnt.plot.bar(rot=50, figsize=(15, 3))
plt.show()
```

```python
# 결측치를 가로 막대그래프(barh)로 시각화
# missing_cnt.plot(kind='barh')
missing_cnt.plot.barh(figsize=(10, 7))
plt.show()
```

```python
missing_cnt_sorted = missing_cnt.sort_values(ascending=False)  # 내림차순(descending)
```

```python
missing_cnt_sorted.plot(kind='bar', rot=50, figsize=(15, 3))
plt.show()
```

```python
missing_cnt.sort_values().plot(kind='barh', figsize=(10, 8))
plt.show()
```

## ④ 결측치가 없는 컬럼는 제외하고 시각화(정렬전 bar plot, barh plot, 정렬 후 bar plot, barh plot)

```python
# missing_cnt  # 열별 결측치 갯수
only_missing_cnt = missing_cnt[missing_cnt != 0]
only_missing_cnt
```

```python
# only_missing_cnt(결측치가 없는 열 갯수)를 barplot으로 그리기
only_missing_cnt.plot.bar(figsize=(15, 5), rot=0 , ylim=[0, 700000])
for i, v in enumerate(only_missing_cnt):
    if v < 10:
        plt.text(i-0.05, v+20000, v)
    elif v < 1000:
        plt.text(i-0.1, v+20000, v)
    else:
        plt.text(i-0.26, v+20000, v)
plt.show()
```

```python
# only_missing_cnt(결측치가 없는 열 갯수)를 barplot으로 그리기
plt.figure(figsize=(15, 5))
g = sns.barplot(data=only_missing_cnt.to_frame(),
                x=only_missing_cnt.index, y=only_missing_cnt)
for i, v in enumerate(only_missing_cnt):
    if v < 10:
        plt.text(i-0.05, v+5000, v, color='r')
    elif v < 1000:
        plt.text(i-0.1, v+5000, v, color='b', fontsize=15)
    else:
        plt.text(i-0.26, v+5000, v)
# plt.title('결측치 갯수')
# g.set(title='결측치 갯수')
g.set_title('결측치 갯수', color='g', weight='bold')
plt.show()
```

```python
only_missing_cnt_sorted = only_missing_cnt.sort_values(ascending=False)
only_missing_cnt_sorted
```

```python
only_missing_cnt_sorted.plot.bar(ylim=[0, 650000], yticks=[0, 100000, 200000, 400000, 500000, 550000, 600000, 650000], 
                                 figsize=(15, 5), rot=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)  # y축만 투명도 0.5인 점선 grid 추가
for i, v in enumerate(only_missing_cnt_sorted):
    if v < 10:
        plt.text(i-0.05, v+5000, v)
    elif v < 100:
        plt.text(i-0.1, v+5000, v)
    elif v < 1000:
        plt.text(i-0.15, v+5000, v)
    else:
        plt.text(i-0.32, v+5000, v, color='r', fontweight='bold', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
g = sns.barplot(data=only_missing_cnt_sorted.to_frame(),
                x=only_missing_cnt_sorted, y=only_missing_cnt_sorted.index)
g.set_xlim([0, 700000])
g.set_xticks([0, 100000, 200000, 400000, 500000, 550000, 600000, 650000, 700000])
plt.grid(axis='x', linestyle='--', alpha=0.5)  # x축만 투명도 0.5인 점선 grid 추가
for i, v in enumerate(only_missing_cnt_sorted):
    if v < 1000:
        plt.text(v+5000, i+0.2, v)
    else:
        plt.text(v+5000, i+0.2, v, color='r', fontweight='bold', fontsize=12)
plt.show()
```

# 4. df 데이터 셋에서 사용하지 않을 컬럼은 제거

## ① 결측치가 너무 많은 컬럼은 제거한다. 결측치가 너무 많은 상위 9개 컬럼은 제거함.

```python
# 결측치가 많은 컬럼 9개와 결측치 갯수
not_use = df.isnull().sum().sort_values(ascending=False).head(9)
not_use
```

```python
df.shape, len(df)
```

```python
# 호정보 컬럼들의 결측치 비율
not_use['호정보'] / df.shape[0] * 100
```

```python
# not_use에 있는 컬럼들의 결측치 비율
round(not_use / df.shape[0] * 100, 2)
```

```python
not_use.index
```

```python
# not_use에 있는 컬럼들의 결측치 비율(표준산업분류명 0.02%) - 리스트컴프리헨션 사용
for col, percent in zip(not_use.index, [str(round(x / len(df)*100, 2))+'%' for x in not_use]):
    print(col, percent)
```

```python
# not_use에 있는 컬럼들의 결측치 비율(표준산업분류명 0.02%) - map 사용
not_use.map(lambda x : str(round(x / len(df)*100, 2))+'%')
```

```python
# not_use에 있는 컬럼들의 결측치 비율(표준산업분류명 0.02%) - apply 사용
not_use.apply(lambda x : str(round(x / len(df)*100, 2))+'%')
```

```python
# 삭제할 컬럼명
not_use.index
```

```python
df.info()
```

```python
# 컬럼 삭제 전 메모리 사용량(181.5+ MB, 컬럼 수 확인
print('결측치 컬럼 제거 전 shape : ', df.shape)
# 컬럼 삭제
df.drop(not_use.index, axis=1, inplace=True)
# 컬럼 삭제 후 메모리 사용량(139.6+ MB), 컬럼 수 확인
print('결측치 컬럼 제거 후 shape : ', df.shape)
```

```python
df.info()
```

## ② 컬럼명에 “코드”나 “번호”가 있는 컬럼은 분석에 사용되지 않을 예정입니다. 제거함

```python
# 컬럼명
cols = df.columns
cols
```

```python
# cols에 '코드'나 '번호'가 포함되어 있는 컬럼명 추출
cols[(cols.str.find('코드') != -1) | (cols.str.find('번호') != -1)]
```

**※ 시리즈에 문자함수를 쓰기 위한 참조 :
https://pandas.pydata.org/pandas-docs/stable/reference/series.html#string-handling**

```python
drop_cols = cols[cols.str.contains('코드|번호')]  # contains() : 코드나 번호가 포함되어 있으면 True
drop_cols
```

```python
print('코드 및 번호가 포함된 컬럼 제거 전 shape : ', df.shape)
df.drop(drop_cols, axis=1, inplace=True)
print('코드 및 번호가 포함된 컬럼 제거 후 shape : ', df.shape)
```
