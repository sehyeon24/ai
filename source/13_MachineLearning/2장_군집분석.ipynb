# 2절. K-means 클러스터링
- 가장 단순하고 빠른 클러스터링 알고리즘
- 몇 개의 클러스터로 나눌지 미리 지정해야 함

```python
# 그래프 저장 시 해상도 높게
%config InlineBackend.figure_format = 'retina'

# 패키지 import
import matplotlib.pyplot as plt
import seaborn as sns

# 한글설정
plt.rc('font', family='Malgun Gothic')  # 윈도우즈
# plt.rc('font', family='AppleGothic') # Mac 운영체제
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rc('axes', unicode_minus=False)  # 축의 -(마이너스) 깨짐 방지
```

```python
# 분류를 위한 가상 데이터 셋 생성
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=20,           # 데이터 갯수(기본값:100)
                           n_features=2,           # 독립변수 갯수(기본값:20)
                           n_informative=2,        # 독립변수에서 종속변수에 영향을 미치는 변수 갯수
                           n_redundant=0,          # 종속변수에 영향을 미치지 않는 변수 갯수
                           n_clusters_per_class=1, # 각 클래스(그룹)당 군의 갯수(군에 서브그룹이 없음)
                           n_classes=2,            # 종속변수의 클래스 수(2그룹으로 나뉨)
                           random_state=123)
X.shape, y.shape
```

```python
import numpy as np
# np.c_ : 열별 병합
np.c_[X, y][:3]
```

```python
plt.figure(figsize=(8, 3))
plt.scatter(x=X[:, 0], y=X[:, 1], c=y, s=100)  # c:y별 다른 색상, s:점의 크기
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2,  # 클러스터 갯수:2개의 그룹으로 나눔
               init='random', # random(중심초기점이 random), k-means++(멀리 떨어진 초기점)
               n_init=5,
               max_iter=300)  # 300번 중심점 이동
model.fit(X)
```

```python
model.cluster_centers_  # 최종 두 클러스터의 중심점
```

```python
pred = model.predict(X)
print(' 예 측 값 : ', pred)
print('modelLabel: ', model.labels_)
print(' 실제 y값 : ', y)
```

```python
# k-means가 0그룹으로 분류된 X 데이터
X[model.labels_ == 0]
```

```python
c0, c1 = model.cluster_centers_  # 중심점 2개
plt.figure(figsize=(10, 3))
plt.scatter(x=X[:, 0], y=X[:, 1], marker='x', c=y, s=200, linewidth=3)  # 실제값
plt.scatter(x=X[model.labels_ == 0, 0], y=X[model.labels_ == 0, 1], marker='o', c='b', s=100)
plt.scatter(x=X[model.labels_ == 1, 0], y=X[model.labels_ == 1, 1], marker='o', c='g', s=100)
plt.scatter(x=c0[0], y=c0[1], marker='s', c='r', s=150)
plt.scatter(x=c1[0], y=c1[1], marker='s', c='r', s=150)
plt.show()
```

```python
# k-means가 1그룹으로 분류된 X 데이터
X[model.labels_ == 1]
```

```python
import pandas as pd
pd.crosstab(y, model.labels_, rownames=['실제값'], colnames=['k-means값'])
```

```python
# 거리 계산 시 스케일 조정
from sklearn.preprocessing import Normalizer
scaler = Normalizer()
Xn = scaler.fit_transform(X)
X_df = pd.DataFrame(X, columns=['x0', 'x1'])
Xn_df = pd.DataFrame(Xn, columns=['x0_nor', 'x1_nor'])
pd.concat([X_df, Xn_df], axis=1)
```

```python
model = KMeans(n_clusters=2, init='k-means++', n_init=5, max_iter=300)
model.fit(Xn)
print('그룹결과 : ', model.labels_)
```

```python
c0, c1 = model.cluster_centers_  # 중심점 2개
plt.figure(figsize=(10, 3))
plt.scatter(x=Xn[:, 0], y=Xn[:, 1], marker='x', c=y, s=200, linewidth=3)  # 실제값
plt.scatter(x=Xn[model.labels_ == 0, 0], y=Xn[model.labels_ == 0, 1], marker='o', c='b', s=100)
plt.scatter(x=Xn[model.labels_ == 1, 0], y=Xn[model.labels_ == 1, 1], marker='o', c='g', s=100)
plt.scatter(x=c0[0], y=c0[1], marker='s', c='r', s=150)
plt.scatter(x=c1[0], y=c1[1], marker='s', c='r', s=150)
plt.show()
```

```python
# 회차별 군집 확인하기
def plot_cluster(model, data):
    c0, c1 = model.cluster_centers_  # 중심점 2개
    plt.scatter(x=data[model.labels_ == 0, 0], y=data[model.labels_ == 0, 1], marker='o', c='b', s=100)
    plt.scatter(x=data[model.labels_ == 1, 0], y=data[model.labels_ == 1, 1], marker='o', c='g', s=100)
    plt.scatter(x=c0[0], y=c0[1], marker='s', c='r', s=150)
    plt.scatter(x=c1[0], y=c1[1], marker='s', c='r', s=150)
    return c0, c1
```

```python
plt.figure(figsize=(8, 8))
for i in range(1, 7):
    model = KMeans(n_clusters=2, init='random', n_init=2, max_iter=i, random_state=1)
    model.fit(X)
    plt.subplot(3, 2, i)
    c0, c1 = plot_cluster(model, X)
    print('c0({:.2f}, {:.2f}), c1({:.2f}, {:.2f})'.format(c0[0], c0[1], c1[0], c1[1]))
plt.show()
```

# 3절. Hierarchical 클러스터링 (계층적 군집분석)
- 군끼리 묶으며 최종적으로 하나의 군집이 될 때까지 클러스터링하는 알고리즘
- dendrogram이라는 그래프를 이용하여 시각화
- 군의 갯수를 미리 정하지 않음

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.sample()
```

```python
# 계층적 군집분석을 위해 모든 데이터 수치화 (species 라벨인코딩)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris.species)
iris.iloc[::50]
```

```python
from scipy.cluster.hierarchy import linkage
cluster_model = linkage(iris, method='complete')
# method='complete' : 두 그룹의 특성 차이를 거리로 측정
```

```python
# 형성된 계층적 모델로 dendrogram 표현
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(20, 5))
dendrogram(cluster_model, labels=iris.index)
# 그림파일로 저장
plt.savefig('ch2_dendrogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# y축 값 기준으로 몇 개의 그룹으로 나뉘는지 확인
# 임계값이 4일 때 그룹
from scipy.cluster.hierarchy import fcluster
fcluster(cluster_model, t=4, criterion='distance')
```

```python
# 임계값이 6일 때 그룹
fcluster(cluster_model, t=6, criterion='distance')
```

```python
# 임계값이 2일 때 그룹
fcluster(cluster_model, t=2, criterion='distance')
```

# 4절. DBSCAN 클러스터링
- 연결기반 군집화 알고리즘
- 클러스터의 수를 미리 정할 필요가 없음
- 이상치 자동 감지
- eps(반경)과 min_sample(최소 샘플수)을 적절히 조정해야 함

```python
# rand(size) : 균등분포인 size개 난수 발생
X1 = np.random.rand(200)*3
noise = np.random.normal(0, 0.2, X1.shape[0])  # 평균0, 표준편차0.2, X1.shape개
Y1 = np.sin(X1) + noise
group1 = np.c_[X1, Y1, np.full(X1.shape[0], 0)]

X2 = X1 + 1.5
Y2 = np.cos(X2) + noise
group2 = np.c_[X2, Y2, np.full(X2.shape[0], 1)]

data = np.r_[group1, group2]
plt.scatter(x=data[:, 0], y=data[:, 1], c=data[:, 2])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
# 반경 0.3 이내의 포인트들을 같은 그룹으로 묶음
db_model = DBSCAN(eps=0.3, min_samples=10)
db_model.fit(data[:, 0:2])
```

```python
labels = db_model.labels_
# labels -1 : 이상치
set(labels)
```

```python
colors = ['r', 'g']
for i, l in enumerate(set(labels)):
    # print(i, l)
    plt.scatter(x=data[labels == i, 0], y=data[labels == i, 1], c=colors[i])
plt.scatter(x=data[labels == -1, 0], y=data[labels == -1, 1], s=100, marker='x', c='k')  # 이상치 표시
plt.show()
```

## 5.2 실루엣 계수
- 클러스터 소속과 갯수를 모를 때 평가하는 방법

```python
from sklearn.metrics import silhouette_score
silhouette_score(iris_X, model3.labels_)
```

```python
X1 = np.random.rand(200)*3
noise = np.random.normal(0, 0.2, X1.shape[0])  # 평균0, 표준편차0.2, X1.shape개
Y1 = np.sin(X1) + noise
group1 = np.c_[X1, Y1]

X2 = X1 + 1.5
Y2 = np.cos(X2) + noise
group2 = np.c_[X2, Y2]

data = np.r_[group1, group2]
plt.scatter(x=data[:, 0], y=data[:, 1])
plt.show()
```

```python
# 실루엣 계수를 이용해 몇 개의 그룹으로 클러스터링 하는 것이 가장 좋은지 판별
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for n_clusters in range_n_clusters:
    model = KMeans(n_clusters=n_clusters, n_init=10, random_state=2)
    model.fit(data)
    s_score = silhouette_score(data, model.labels_)
    print('클러스터 수 : {}일 때, 실루엣 계수 : {}'.format(n_clusters, s_score))
```

```python
# 1. iris 데이터의 petal_length열과 petal_width열을 이용해서 K-Means 알고리즘으로 군집분석하고 그래프로 시각화하세요(단, 각 클러스터의 중심점이 함께 표시되고 군의 수는 2로 설정)
X = iris[['petal_length', 'petal_width']].values  # 독립변수를 array로 하지 않으면 경고메세지 출력

model = KMeans(n_clusters=2,  # 클러스터 갯수:2개의 그룹으로 나눔
               init='random', # random(중심초기점이 random), k-means++(멀리 떨어진 초기점)
               n_init=5,
               max_iter=300)  # 300번 중심점 이동
model.fit(X)
# model = KMeans(n_clusters=2, init='random', n_init=5, max_iter=300).fit(X)

model.cluster_centers_  # 최종 두 클러스터의 중심점

pred = model.predict(X)

plt.figure(figsize=(8, 3))
plt.scatter(x=X['petal_length'], y=X['petal_width'], c=pred, s=50)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='red', s=200, marker='X')
plt.show()
```

```python
# 실루엣 계수(평가 지표) - 2개의 클러스터로 군집분석 시 평가지표
from sklearn.metrics import silhouette_score
silhouette_score(X, model.labels_)
```

```python
# 2. 클러스터 갯수가 3일 때
data = iris[['petal_length','petal_width']].values
model3 = KMeans(n_clusters=3, init='random', n_init=5).fit(data)

# 시각화
centers = model3.cluster_centers_ # 중심점 3개
plt.figure(figsize=(12,3))
plt.scatter(x=data[model3.labels_==0, 0],
            y=data[model3.labels_==0, 1], marker='o', c='b', s=100)
plt.scatter(x=data[model3.labels_==1, 0],
            y=data[model3.labels_==1, 1], marker='o', c='g', s=100)
plt.scatter(x=data[model3.labels_==2, 0],
            y=data[model3.labels_==2, 1], marker='^', c='k', s=100)
plt.scatter(x=centers[:,0], 
            y=centers[:, 1], marker='s', c='r', s=150)
plt.show()

# 실루엣 계수
from sklearn.metrics import silhouette_score
silhouette_score(data, model3.labels_)
```
